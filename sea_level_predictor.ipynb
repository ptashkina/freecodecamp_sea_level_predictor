```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Read data from file
df = pd.read_csv('epa-sea-level.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Create scatter plot
fig, ax = plt.subplots(figsize=(12, 5))
df.plot(kind='scatter', x='Year', y='CSIRO Adjusted Sea Level', ax=ax)
```

```python
slope1, intercept1, r, p, se = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
#predicted
y1 = [slope1 * d + intercept1 for d in range(1880, 2050)]
fig, ax = plt.subplots(figsize=(12, 5))
df.plot(kind='scatter', x='Year', y='CSIRO Adjusted Sea Level', ax=ax)
plt.plot(range(1880, 2050), y1, color='red')
```

```python
df2 = df[df['Year'] >= 2000]
slope2, intercept2, r, p, se = linregress(df2['Year'], df2['CSIRO Adjusted Sea Level'])
```

```python
#line 1 + line 2
y1 = [slope1 * d + intercept1 for d in range(1880, 2050)]
y2 = [slope2 * d + intercept2 for d in range(2000, 2050)]
fig, ax = plt.subplots(figsize=(12, 5))
df.plot(kind='scatter', x='Year', y='CSIRO Adjusted Sea Level', ax=ax)
plt.plot(range(1880, 2050), y1, color='red')
plt.plot(range(2000, 2050), y2, color='green')
ax.set_title('Rise in Sea Level')
ax.set_ylabel('Sea Level (inches)') 
```
